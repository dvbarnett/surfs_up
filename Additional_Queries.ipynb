```python
# Dependencies
import numpy as np
import pandas as pd
import datetime as dt

# Python SQL toolkit and Object Relational Mapper
import sqlalchemy
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, func
```

```python
engine = create_engine("sqlite:///hawaii.sqlite")

# reflect an existing database into a new model
Base = automap_base()
# reflect the tables
Base.prepare(engine, reflect=True)

# Save references to each table
Measurement = Base.classes.measurement
Station = Base.classes.station
```

```python
# Create our session (link) from Python to the DB
session = Session(engine)
```

```python
# 1. Import the sqlalchemy extract function.
from sqlalchemy import extract
```

```python
# 2. Write a query that filters the Measurement table to retrieve the precipitation for the month of June. 
june_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 6).all()
june_prcp
```

```python
june_precip = list(june_prcp)
```

```python
june_precip_df = pd.DataFrame(june_precip, columns= ['date', 'June Precipitation'])
june_precip_df = june_precip_df.set_index(june_precip_df['date'], inplace=False)
print(june_precip_df)
```

```python
june_precip_df.describe()
```

```python
june_precip_df = june_precip_df.sort_index()
```

```python
june_precip_df.plot.scatter('date','June Precipitation')
```

```python
dec_prcp = session.query(Measurement.date, Measurement.prcp).filter(extract('month', Measurement.date) == 12).all()
dec_prcp
```

```python
dec_precip = list(dec_prcp)
```

```python
dec_precip_df = pd.DataFrame(dec_precip, columns= ['date', 'December Precipitation'])
dec_precip_df = dec_precip_df.set_index(dec_precip_df['date'], inplace=False)
print(dec_precip_df)
```

```python
dec_precip_df.describe()
```

```python
dec_precip_df = dec_precip_df.sort_index()
```

```python
dec_precip_df.plot.scatter('date','December Precipitation')
```
